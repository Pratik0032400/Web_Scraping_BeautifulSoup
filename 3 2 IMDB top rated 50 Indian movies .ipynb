```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url= 'https://www.imdb.com/list/ls025646548/'
```

```python
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
a='h3'
b=class_="lister-item-header"
```

```python
n = []
for i in soup.find_all(a,b):
    n.append(i.text.replace('\n','').split('.')[1])
n
```

```python
c='div'
d=class_="ipl-rating-star small"
```

```python
rating = []
for i in soup.find_all(c,d):
    rating.append(i.text.replace('\n',''))
rating
```

```python
le=len(rating)
le
```

```python
e ='span'
f =class_="lister-item-year text-muted unbold"
```

```python
year = []
for i in soup.find_all(e,f):
    year.append(i.text)
year
```

```python
df=pd.DataFrame({'NAME':n,"RATING":rating,"YEAR":year})
df
```
